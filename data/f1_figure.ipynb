## DHS Educational Attainment and Health Insurance Data Exploration

## By John Adeleke

```stata
*ssc install dubois
```

```stata
***load in the saved dataset

use dhswomenrecode_nigeria.dta, clear
```

```stata
qui reg health_insurance i.education, ro

qui margins i.education, atmeans 
```

```stata
marginsplot, scheme(dubois) ///
title("Figure 1: Health Insurance by Educational attainment" " ") ///
    ytitle(proportion with health insurance) 
```

```stata
qui reg health_insurance i.residence, ro

qui margins i.residence, atmeans 
```

```stata
marginsplot, scheme(dubois) ///
title("Figure 1: Health Insurance by Urban and Rural Residence" " ") ///
    ytitle(proportion with health insurance) 
```

```stata
qui reg health_insurance i.age, ro

qui margins i.age, atmeans
```

```stata
marginsplot, scheme(dubois) ///
title("Figure 1: Health Insurance by Age of Women" " ") ///
    ytitle(proportion with health insurance) 
```
